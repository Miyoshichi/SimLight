```python
import numpy as np

import SimLight as sl
import SimLight.zernike as slz
import SimLight.plottools as slpl
```

```python
# units
m = 1
cm = 1e-2
mm = 1e-3
µm = 1e-6
nm = 1e-9
# parameters
wavelength = 0.6328 * µm  # wavelength of laser
size = (15 * µm, 15 * µm, 1 * µm)  # size of measured surface
N = (150, 150, 10)  # resolution
n0 = 1.37

np.random.seed(235)
```

```python
dn = np.random.rand(*N)
dn = (dn - 0.5) / 5
```

```python
S = sl.ScatteringLayer.new_scattering_layer(wavelength, size=size, N=N, dn=dn, n0=n0)
```

```python
S.complex_amp.shape
```

```python
slpl.plot_wavefront(S, mask_r=1, unit='µm', dimension=2, title='Phase distribution of scattering tissue')
```

```python
import matplotlib.pyplot as plt
dn_bottom = dn[:, :, -1]
plt.figure()
plt.imshow(dn_bottom, cmap='RdBu', vmin=-0.1, vmax=0.1)
plt.xticks([])
plt.yticks([])
plt.title('$d_n$ distribution')
plt.colorbar(ticks=[-0.1, 0, 0.1])
plt.show()
```
